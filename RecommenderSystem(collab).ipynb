# **DATA PREPROCESSING**

```python
import numpy as np
import pandas as pd
import ast
import nltk
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv', on_bad_lines='skip')
```

```python
#merging dataframes (movies and credits)
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
#selecting columns
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
#converting string to list (GENRES)
movies.iloc[0].genres
def convert (obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
#converting string to list (Keywords)
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
#picking only 3 dictionaries from cast
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
movies['cast']=movies['cast'].apply(convert3)
movies.head()
```

```python
#work on crew and choosing job as director
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
movies['crew']=movies['crew'].apply(fetch_director)
movies.head()
```

```python
#overview in list from string
# Handle potential float values in 'overview' column
def split_overview(x):
  if isinstance(x, str):
    return x.split()
  else:
    return []  # Return an empty list for non-string values

movies['overview'] = movies['overview'].apply(split_overview)
movies.head()
#movies['overview'].apply(lambda x:x.split())
```

```python
#removing spaces between words
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()
```

```python
#creating tag column
movies['tags']=movies['overview']+movies['cast']+movies['crew']+movies['genres']+movies['keywords']
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df
```

```python
#lower case
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df
```

# **VECTORISATION**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
# to change names (like singular , plural, adverbs ) into single rooted word
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):

  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df
```

```python
#finding cosine angles between various vectors and closest
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
#sorted(similarity[0],reverse=True)
def recommend(movie):
  #to fetch index of movies from datafame
  movie_index=new_df[new_df['title'] == movie].index[0]
  distances=similarity[movie_index]
  #writing index with distances(list of tuples)
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
